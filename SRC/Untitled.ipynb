```python
import pylab as plt
import seaborn as sns
import numpy as np 
```

```python
ocde.head()
```

```python
ocde.drop(['Emp.BUPPSRY','Emp.UPPSRY_NTRY','Emp.TRY'], axis = 1, inplace = True)
```

```python
plt.figure(figsize=(15, 10))

sns.set(style='white')

mask=np.triu(np.ones_like(ocde.corr(), dtype=bool))

cmap=sns.color_palette("ch:start=.2,rot=-.3", as_cmap=True)


sns.heatmap(ocde.corr(),
           mask=mask,
            annot=True,
          cmap=cmap,
          center=0,
          square=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});
plt.savefig("foo.png", dpi=100)
```
